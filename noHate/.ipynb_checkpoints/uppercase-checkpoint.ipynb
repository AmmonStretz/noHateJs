```python
start = 1
stop = 16
step = 16
# the filter of the table
def uppercaseWords(text):
    counter = 0
    for word in text:
        if word.isupper():
            counter += 1;
    return counter
pfilters = []
def generate_filter(fr, to):
    def f(element):
        ld = uppercaseWords(element['tokens'])
        if ld >= fr and ld < to:
            return True
        return False
    return f
pfilters.append(('all',generate_filter(0, 2000)))
for i in range(start, stop, step):
    pfilters.append(('%s-%s' % (i, i+step-1),generate_filter(i, i+step)))

for t in data:
    print(uppercaseWords(t['tokens']))
# print(pfilters)
```

```python
#name, filters, relative
createPlt(name = "Test1 absolute", filters = pfilters, relative = False)
createPlt(name = "Test1 relative", filters = pfilters, relative = True)
```
